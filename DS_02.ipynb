## Question 1: Extracting Tesla Stock Data Using yfinance

```python
import yfinance as yf
import pandas as pd

# Get Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Reset index to move Date to a column
tesla_data.reset_index(inplace=True)

# Show first 5 rows
tesla_data.head()
```

## Question 2: Extracting Tesla Revenue Data Using Web Scraping

```python
# Load Tesla revenue from file
tesla_revenue = pd.read_csv("tesla_revenue.csv")
tesla_revenue["Date"] = pd.to_datetime(tesla_revenue["Date"])

# Display first few rows
tesla_revenue.head()
```

## Question 3: Extracting GameStop Stock Data Using yfinance

```python
# Get GameStop stock data
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")
gme_data.reset_index(inplace=True)

# Display first few rows
gme_data.head()
```

## Question 4: Extracting GameStop Revenue Data Using Web Scraping

```python
# Load GameStop revenue from file
gme_revenue = pd.read_csv("gamestop_revenue.csv")
gme_revenue["Date"] = pd.to_datetime(gme_revenue["Date"])

# Display first few rows
gme_revenue.head()
```

## Question 5: Tesla Stock and Revenue Dashboard

```python
import matplotlib.pyplot as plt

tesla_revenue["Date"] = pd.to_datetime(tesla_revenue["Date"])
tesla_data["Date"] = pd.to_datetime(tesla_data["Date"])

plt.figure(figsize=(14, 6))
plt.plot(tesla_data["Date"], tesla_data["Close"], label="Tesla Stock Price")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.title("Tesla Stock Price Over Time")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(14, 6))
plt.plot(tesla_revenue["Date"], tesla_revenue["Revenue"], color='orange', label="Tesla Revenue")
plt.xlabel("Date")
plt.ylabel("Revenue (Millions USD)")
plt.title("Tesla Quarterly Revenue")
plt.legend()
plt.grid(True)
plt.show()
```

## Question 6: GameStop Stock and Revenue Dashboard

```python
gme_revenue["Date"] = pd.to_datetime(gme_revenue["Date"])
gme_data["Date"] = pd.to_datetime(gme_data["Date"])

plt.figure(figsize=(14, 6))
plt.plot(gme_data["Date"], gme_data["Close"], label="GameStop Stock Price", color="green")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.title("GameStop Stock Price Over Time")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(14, 6))
plt.plot(gme_revenue["Date"], gme_revenue["Revenue"], color='purple', label="GameStop Revenue")
plt.xlabel("Date")
plt.ylabel("Revenue (Millions USD)")
plt.title("GameStop Quarterly Revenue")
plt.legend()
plt.grid(True)
plt.show()
```
